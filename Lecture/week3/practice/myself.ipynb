__1.8__. Write code to load 'test_image' using the same method as before and isolate the grass of the image by filter out values within the blue channel that are below 100. Then reconstruct the image(by multipliying the original values by the filered ones) and plot it.

```python
import numpy as np
import matplotlib.pylab as plt

test = plt.imread("test_image.jpg")
plt.imshow(test)
```

```python
#test1 is used to filted
test1 = test.copy()
plt.imshow(test1)
#filtered test1
test2=test1[:,:,2]<100
plt.imshow(test2)
```

```python
#then reconstruct the image(by multipliying the original values by the filered ones) 
test_new = test1.copy()
test_new[:,:,0] = test2*test_new[:,:,0]
test_new[:,:,1] = test2*test_new[:,:,1]
test_new[:,:,2] = test2*test_new[:,:,2]
plt.imshow(test_new)
```

__1.9__. Write code to compute the coordinates for points on a cosine curve and plot the points using matplotlib __(Optional)__

Reference: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Compute the x and y coordinates for points on a cosine curve
x = np.arange(0, 3 * np.pi, 0.02)
y = np.cos(x)
print("Plot the points using matplotlib:")
plt.plot(x, y)
plt.show()
```
